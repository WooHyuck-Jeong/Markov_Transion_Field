```python
import numpy as np
import pandas as pd
import os
```

```python
# npz 파일 중 i_data 불러오기

Welding_data = np.load('E:/3.08/MTF/MTF.npz')

i_data = Welding_data['i_data']

Welding_data.close()
```

```python
# i_data csv save

np.savetxt('E:/3.08/MTF/i_data/i_data_10.csv', i_data, fmt='%s', header='i_data')
```

```python
# index = i_data

i_data = pd.read_csv('E:/3.08/MTF/i_data/i_data_10.csv')
i_data = np.array(i_data)
i_data = i_data.reshape(len(i_data),)
print(i_data.shape)
```

```python
print(i_data)
```

```python
# i_data_list : i_data의 순서
i_data_list = []

for i in range(len(i_data)):
    i_data_list.append(i)
```

```python
# key : i_data_list
# value : i_data

diction_i_data = dict(zip(i_data_list, i_data))

for key, element in diction_i_data.items():
    print("{} = {}".format(key, element))
```

```python
# train_data.csv 불러오기

# trainset = train

trainset = pd.read_csv('E:/3.08/MTF/train/train_10.csv')
trainset = np.array(trainset)
trainset = trainset.reshape(len(trainset),)
print(trainset.shape)
print(trainset)
```

```python
# diction_i_data의 value값을 train_list에 저장

train_list = []

for i in range(0,len(trainset)):
    train_list.append(diction_i_data.get(trainset[i]))
    print(diction_i_data.get(trainset[i]))
```

```python
# train_list.csv로 저장
np.savetxt('E:/3.08/MTF/train_list/train_list_10.csv', train_list, fmt="%s", header='train_list')
```

```python
# Image dictionary

# Image index 생성

index_length = int(len(i_data)/3)
print(index_length)

# Humping index
list_Humping_index = []
for i in range(index_length):
    list_Humping_index.append('H_' + str(i))

# Normal index
list_Normal_index = []
for j in range(index_length):
    list_Normal_index.append('N_'+str(j))

# Porosity index
list_Porosity_index = []
for k in range(index_length):
    list_Porosity_index.append('P_'+str(k))

Humping_index = np.array(list_Humping_index)
Normal_index = np.array(list_Normal_index)
Porosity_index = np.array(list_Porosity_index)
```

```python
# N,H,P index concatenate

Merge_index = np.concatenate((Humping_index, Normal_index, Porosity_index), axis=0)

print(Merge_index.shape)
```

```python
# Merge_index.csv 저장
np.savetxt('E:/3.08/MTF/Merge_index/Merge_index_10.csv', Merge_index, fmt="%s")
```

```python
# Image import

from PIL import Image

list_Merge_Humping = []
list_Merge_Normal = []
list_Merge_Porosity = []

for i in range(0,len(list_Normal_index)):
    M_H = Image.open('E:/3.08/MTF/Image/Merge_Humping/' + f'H_{i}.tiff')
    M_N = Image.open('E:/3.08/MTF/Image/Merge_Normal/' + f'N_{i}.tiff')
    M_P = Image.open('E:/3.08/MTF/Image/Merge_Porosity/' + f'P_{i}.tiff')

    M_H = np.array(M_H)
    M_H = M_H.reshape(200,200,2)

    M_N = np.array(M_N)
    M_N = M_N.reshape(200,200,2)

    M_P = np.array(M_P)
    M_P = M_P.reshape(200,200,2)

    list_Merge_Humping.append(M_H)
    list_Merge_Normal.append(M_N)
    list_Merge_Porosity.append(M_P)
```

```python
M_H = np.array(list_Merge_Humping)
M_N = np.array(list_Merge_Normal)
M_P = np.array(list_Merge_Porosity)
```

```python
# Image list concatenate

Merge_image = np.concatenate((M_H,M_N,M_P), axis=0)

print(Merge_image.shape)
```

```python
# key : Merge_index
# value : Merge_image

diction_image = dict(zip(Merge_index, Merge_image))
```

```python
for key,element in diction_image.items():
    print("{} = {}".format(key,element))
```

```python
# diction_image key에 diction_i_data value 넣어서 학습에 사용 된 이미지만 따로 저장


train_image_list = []

for i in range(len(train_list)):
    train_image_list.append(diction_image.get(train_list[i]))
```

```python
train_image = np.array(train_image_list)
print(train_image.shape)
```

```python
for i in range(len(train_list)):
    print(train_list[i])
```
